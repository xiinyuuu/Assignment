```python
#SIF3012 Tuesday 10/12/2024

# Import libraries
from scipy.fftpack import fft

import matplotlib.pyplot as mpl
import numpy as np

# Number of sample points
N = 1000

# Sample spacing
T = 1.0/800.0

x = np.linspace(0.0, N*T, N)
y = np.sin(50.0*np.pi*x) + 0.5*np.sin(80.0*2.0*np.pi*x)

# Perform dft by calling the scipy library
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)

# Perform dft using numpy library
ynp = np.fft.fft(y)

titles = ['x', 'scipy', 'numpy']
data = [titles]+list(zip(xf, yf, ynp))

# Print results
print ('Result from scipy and numpy')
for i,d in enumerate(data):
    line = '|'.join(str(x).ljust(12) for x in d)
    # Print (line)

# Plot result
mpl.title ('$f(x)=\sin(50(2\pi x) + 0.5\sin(80(2\pi x)))$')
mpl.xlabel('x')
mpl.ylabel('f(x)')
mpl.plot(x,y, label='original function')

#mpl.plot(x, yf, color='k', label='scipy fft')
#mpl.plot(x, ynp, color='r', linestyle=':', label='numpy fft')

#mpl.plot(xf, 2.0/N*abs(yf[0:N//2]), color='k', label='scipy fft')
#mpl.plot(xf, 2.0/N*abs(ynp[0:N//2]), color='r', label='numpy fft')
```
